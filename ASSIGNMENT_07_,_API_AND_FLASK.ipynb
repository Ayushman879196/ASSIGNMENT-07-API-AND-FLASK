 Q1. What is a RESTful API?

 Ans: A RESTful API is an application programming interface that follows the principles of REST
 (Representational State Transfer). It uses standard HTTP methods (GET, POST, PUT, DELETE)
 for communication and is stateless, meaning each request is independent.

 Q2. Explain the concept of API specification

 Ans: An API specification defines how an API should behave. It includes details such as endpoints,
 request/response formats, methods allowed, and authentication mechanisms. Examples include
 OpenAPI (Swagger) specifications.

 Q3. What is Flask, and why is it popular for building APIs

 Ans: Flask is a lightweight Python web framework. It is popular because it is simple, flexible, easy
 to learn, and allows developers to quickly build APIs.

 Q4. What is routing in Flask

 Ans: Routing is the process of mapping URLs to specific functions in Flask. It determines what
 code runs when a specific endpoint is accessed.

 Q5. How do you create a simple Flask application

 Ans: Install Flask, create a Python file, import Flask, create an app instance, define routes, and run
 the app using app.run().

 Q6. What are HTTP methods used in RESTful APIs

 Ans: The main HTTP methods are GET (retrieve data), POST (create data), PUT (update data),
 DELETE (remove data), and PATCH (partial update).

 Q7. What is the purpose of the @app.route() decorator in Flask

 Ans: It is used to bind a URL endpoint to a function. When the endpoint is accessed, the function
 executes.

 Q8. What is the difference between GET and POST HTTP methods

 Ans: GET is used to request data and does not modify resources, while POST is used to submit
 data to create or modify resources.

 Q9. How do you handle errors in Flask APIs

 Ans: Use Flask's errorhandler decorators (e.g., @app.errorhandler(404)) or return custom JSON
 responses with appropriate HTTP status codes.

 Q10. How do you connect Flask to a SQL database

 Ans: You can use libraries like Flask-SQLAlchemy to connect and interact with databases in Flask.

 Q11. What is the role of Flask-SQLAlchemy

 Ans: It provides ORM (Object Relational Mapping) support, making it easier to interact with
 databases using Python classes instead of raw SQL.

 Q12. What are Flask blueprints, and how are they useful

 Ans: Blueprints allow modular structuring of Flask applications. They help organize code into
 components, making apps easier to maintain.

 Q13. What is the purpose of Flask's request object

 Ans: The request object stores data about the incoming request such as form data, query
 parameters, headers, and JSON payload.

 Q14. How do you create a RESTful API endpoint using Flask

 Ans: Define a route using @app.route(), specify the HTTP me
 thod, and return JSON responses
 using jsonify().

 Q15. What is the purpose of Flask's jsonify() function

 Ans: jsonify() converts Python dictionaries into JSON responses with the correct MIME type
 (application/json).

 Q16. Explain Flask’s url_for() function

 Ans: url_for() generates dynamic URLs for routes based on function names, helping avoid
 hardcoding of URLs.

 Q17. How does Flask handle static files (CSS, JavaScript, etc.)

 Ans: Flask serves static files from the 'static' directory. You can access them via the /static/ path.

 Q18. What is an API specification, and how does it help in building a Flask API

 Ans: It provides a contract for API behavior, ensuring consistency and easy collaboration between
 frontend and backend developers.

 Q19. What are HTTP status codes, and why are they important in a Flask API

 Ans: HTTP status codes indicate the result of an API request (e.g., 200 OK, 404 Not Found, 500
 Internal Server Error). They improve API clarity.

 Q20. How do you handle POST requests in Flask

 Ans: Use request.form or request.json to extract data, process it, and return a response.

 Q21. How would you secure a Flask API

 Ans: Use authentication (JWT, OAuth), HTTPS, input validation, and proper error handling.

 Q22. What is the significance of the Flask-RESTful extension

 Ans: Flask-RESTful simplifies building REST APIs by providing tools for request parsing, resource
 routing, and response formatting.

 Q23. What is the role of Flask’s session object?

 Ans: The session object stores user-specific data across requests, such as login information, using
 secure cookies.

 Q24. How do you create a basic Flask application

 Ans: from flask import Flask app = Flask(__name__) @app.route('/') def home(): return 'Hello,
 Flask!' app.run()

 Q25. How do you serve static files like images or CSS in Flask

 Ans: Place files in the static folder and access them via /static/filename.

 Q26. How do you define different routes with different HTTP methods in Flask

 Ans: Use @app.route('/path', methods=['GET','POST']).

 Q27. How do you render HTML templates in Flask

 Ans: Use Flask's render_template() function with templates stored in the 'templates' folder.

 Q28. How can you generate URLs for routes in Flask using url_for

 Ans: Use url_for('function_name') to dynamically generate URLs.

 Q29. How do you handle forms in Flask

 Ans: Access data via request.form and validate it before processing.

 Q30. How can you validate form data in Flask

 Ans: Use WTForms or manual validation (checking required fields, data types, etc.).

 Q31. How do you manage sessions in Flask

 Ans: Use the session object to store and retrieve user-specific data across requests.

 Q32. How do you redirect to a different route in Flask

 Ans: Use Flask's redirect() function along with url_for().

 Q33. How do you handle errors in Flask (e.g., 404)

 Ans: Define custom error handlers with @app.errorhandler(404).

 Q34. How do you structure a Flask app using Blueprints

 Ans: Organize routes, templates, and static files into separate blueprint modules and register them
 in the main app.

 Q35. How do you define a custom Jinja filter in Flask

 Ans: Use app.template_filter() decorator to define a function as a custom filter.

 Q36. How can you redirect with query parameters in Flask

 Ans: Use redirect(url_for('route', param=value)).

 Q37. How do you return JSON responses in Flask

 Ans: Use jsonify() to send JSON data with the correct content type.

 Q38. How do you capture URL parameters in Flask?

 Ans: Define routes with placeholders (e.g., /user/) and access them as function arguments
